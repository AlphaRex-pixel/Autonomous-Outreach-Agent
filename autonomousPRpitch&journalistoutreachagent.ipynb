```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from openai.types.responses import ResponseTextDeltaEvent
```

```python
load_dotenv(override=True)
```

```python
orchestration_instructions = 'You are an agent who is tasked with researching about a product.' \
' You will be assigned to build an outreach strategy to pitch to journalists and media outlets.'

research_instructions = 'You will tasked with scraping the web to find information about a product.' \
'You will also find the right journalists writing abou the specific industry niche.'

copywriter_instructions = 'You will be tasked with writing a compelling pitch to jounalists and media outlets.'
```

```python
orchestration_agent = Agent(
    name='Orchestration Agent',
    instructions=orchestration_instructions,
    model='gpt-4o-mini'
)

research_agent = Agent(
    name='Research Agent',
    instructions=research_instructions,
    model='gpt-4o-mini'
)

copywriter_agent = Agent(
    name='Copywriter Agent',
    instructions=copywriter_instructions,
    model='gpt-4o-mini'
)
```

```python
result = Runner.run_streamed(orchestration_agent, input='Please create a pitch strategy for an outreach campaign for a vegan ice cream brand')
async for event in result.stream_events():
    if event.type == 'raw_response_event' and isinstance(event, ResponseTextDeltaEvent):
        print(event.delta, end='', flush=True)

print(result.final_output)
```

```python
research_agent_result = Runner.run_streamed(research_agent, input='Please research the vegan ice cream brand and find relevant journalists and media outlets for outreach.')
async for event in research_agent_result.stream_events():
    if event.type == 'raw_response_event' and isinstance(event, ResponseTextDeltaEvent):
        print(event.delta, end='', flush=True)

print(research_agent_result.final_output)
```

```python
copywriter_agent_result = Runner.run_streamed(copywriter_agent, input='Please write a compelling pitch to journalists and media outlets for the vegan ice cream brand based on the research findings.')
async for event in copywriter_agent_result.stream_events():
    if event.type == 'raw_response_event' and isinstance(event, ResponseTextDeltaEvent):
        print(event.delta, end='', flush=True)

print(copywriter_agent_result.final_output)
```

```python
orchestration_description = 'Please create a pitch strategy for an outreach campaign for a vegan ice cream brand. The strategy should include the following: 1) Research the vegan ice cream brand and find relevant journalists and media outlets for outreach. 2) Write a compelling pitch to journalists and media outlets based on the research findings.'
research_description = 'Please research the vegan ice cream brand and find relevant journalists and media outlets for outreach.'
copywriter_description = 'Please write a compelling pitch to journalists and media outlets for the vegan ice cream brand based on the research findings.'

tool1 = orchestration_agent.as_tool(tool_name='orchestration_agent', tool_description=orchestration_description)
tool2 = research_agent.as_tool(tool_name='research_agent', tool_description=research_description)
tool3 = copywriter_agent.as_tool(tool_name='copywriter_agent', tool_description=copywriter_description)

tools = [tool1, tool2, tool3]
```

```python

instructions = 'You are an agent who is supposed to create a pitch strategy for an outreach campaign for a vegan ice cream brand.' \
'The strategy should include the following:' \
'1) Research the vegan ice cream brand and find relevant journalists and media outlets for outreach. ' \
'2) Write a compelling pitch to journalists and media outlets based on the research findings.' \
'You must use the tools provided to complete these tasks.' \

outreach_agent = Agent(
    name='Outreach Agent',
    instructions=instructions,
    model='gpt-4o-mini',
    tools=tools)

message = 'Please create a pitch strategy for an outreach campaign for a vegan ice cream brand.'

with trace('Outreach'):
    result = await Runner.run(outreach_agent, input=message)


print(result.final_output)

```
